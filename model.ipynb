```python
def LoosingTeam(row):
    return row['team2'] if row['winner']==row['team1'] else row['team1']

temp['loser']= temp.apply(LoosingTeam,axis=1)
temp.head()
```

```python

team1_wins=[] 
team2_wins=[] 
for i in range(temp.shape[0]):
    t1_win,t2_win = 0,0
    for j in range(i):
        if((temp.at[i,'team1']==temp.at[j,'winner']) and  (temp.at[i,'date']>temp.at[j,'date']) and (temp.at[i,'Season']==temp.at[j,'Season']) ):
            t1_win=t1_win+1
        elif((temp.at[i,'team2']==temp.at[j,'winner'])and(temp.at[i,'date']>temp.at[j,'date']) and (temp.at[i,'Season']==temp.at[j,'Season']) ):
            t2_win=t2_win+1
    team1_wins.append(t1_win)
    team2_wins.append(t2_win)
    # print(temp.at[i,'team1'],count)    
        # print(i,j)    
```

```python
temp['team1_wins']=team1_wins
temp['team2_wins']=team2_wins
```

```python
team1_loss=[] 
team2_loss=[] 
for i in range(temp.shape[0]):
    t1_loss,t2_loss = 0,0
    for j in range(i):
        if((temp.at[i,'team1']==temp.at[j,'loser']) and  (temp.at[i,'date']>temp.at[j,'date']) and (temp.at[i,'Season']==temp.at[j,'Season']) ):
            t1_loss=t1_loss+1
        elif((temp.at[i,'team2']==temp.at[j,'loser']) and  (temp.at[i,'date']>temp.at[j,'date']) and (temp.at[i,'Season']==temp.at[j,'Season']) ):
            t2_loss=t2_loss+1
    team1_loss.append(t1_loss)
    team2_loss.append(t2_loss)
    # print(temp.at[i,'team1'],count)    
        # print(i,j)    
```

```python
temp['team1_losses']=team1_loss
temp['team2_losses']=team2_loss
```

```python
# if(row['team1']=row['toss_winner']):
#     return row['team2']
# else
#     return row['team2']

def batting_team_(row):
    if (row['inning']==1 and row['toss_decision']=="field"):
        if(row['team1']==row['toss_winner']):
            return row['team2']
        else:
            return row['team1']
    elif (row['inning']==1 and row['toss_decision']=="bat"):
        return row['toss_winner']
    elif (row['inning']==2 and row['toss_decision']=="field"):
        return row['toss_winner']
    elif (row['inning']==2 and row['toss_decision']=="bat"):
        if(row['team1']==row['toss_winner']):
            return row['team2']
        else:
            return row['team1']
    
def bowling_team_(row):
    if (row['inning']==1 and row['toss_decision']=="field"):
        return row['toss_winner']
    elif (row['inning']==1 and row['toss_decision']=="bat"):
        if(row['team1']==row['toss_winner']):
            return row['team2']
        else:
            return row['team1']
    elif (row['inning']==2 and row['toss_decision']=="field"):
        if(row['team1']==row['toss_winner']):
            return row['team2']
        else:
            return row['team1']
    elif (row['inning']==2 and row['toss_decision']=="bat"):
        return row['toss_winner']
        
    
        

delivery_df['batting_team_']= delivery_df.apply(batting_team_,axis=1)

delivery_df['bowling_team_']= delivery_df.apply(bowling_team_,axis=1)
```
